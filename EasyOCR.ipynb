```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

## 1. Read in images or video

```python
IMAGE_PATH = '/content/Aadhar-Card-PVC.jpeg'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH)
result
```

#2.Draw Result

```python
for i in result:
  if len(i[-2]) == 14:
    print("The addahr number of given id is: ", i[-2]) 
```

```python
cord = result[-4][0]
```

```python
x_min, y_min = [min(idx) for idx in zip(*cord)]

x_max, y_max = [max(idx) for idx in zip(*cord)]
```

```python
image = cv2.imread('Aadhar-Card-PVC.jpeg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
